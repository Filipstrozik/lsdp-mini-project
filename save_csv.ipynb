```python
from pymongo import MongoClient

client = MongoClient("mongodb://root:example@localhost:27017/")
db = client["reviews_db"]
collection = db["vectorized_reviews"]

# Get all documents from the collection
documents = list(collection.find({}))
print(f"Loaded {len(documents)} documents from collection")
```

```python
import pandas as pd

# Convert the documents list to a DataFrame
df = pd.DataFrame(documents)

# Save the DataFrame to a CSV file
# The encoding 'utf-8-sig' helps preserve special characters
path = "spark/data/vectorized_reviews.csv"
df.to_csv(path, index=False, encoding="utf-8-sig")

# Display the first few rows and shape of the DataFrame
print(f"DataFrame shape: {df.shape}")
df.head()
```
